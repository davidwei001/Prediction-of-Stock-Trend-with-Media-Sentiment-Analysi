## 1. read tweets data

```python
tweets_dataf = pd.read_csv('MSFT_twi08.csv', error_bad_lines=False)
tweets_dataf = tweets_dataf[['Date','Tweet']]
tweets_dataf
```

## 2. comine by date

```python
combine_dataf = pd.DataFrame(columns = ['Date', 'Tweet'])
indx = 0
combine_tweet = ""
for i in range(1, len(tweets_dataf) - 1):##### skip 10-31
    temp_date = tweets_dataf.Date.iloc[i]
    next_date = tweets_dataf.Date.iloc[i+1]
    if (str(temp_date)==str(next_date)):
        combine_tweet = combine_tweet + tweets_dataf.Tweet.iloc[i] + " "
    if (str(temp_date) != str(next_date)):
        combine_dataf.set_value(indx, 'Date', temp_date)
        combine_dataf.set_value(indx, 'Tweet', combine_tweet)
        indx = indx + 1
        combine_tweet = " "
    if (i == len(tweets_dataf) - 2):
        combine_dataf.set_value(indx, 'Date', temp_date)
        combine_dataf.set_value(indx, 'Tweet', combine_tweet)
combine_dataf
```

```python
combine_dataf.to_csv("combine_dataf.csv", sep='\n',index=False,header=False)
```

## 3. read the stock price data

```python
price_dataf = pd.read_csv('MSFT.csv')
price_dataf = price_dataf[['Date', 'Close']]
```

## 4. merge the price and tweets

```python
combine_dataf['Prices'] = ""
indx = 0
for i in range(0, len(combine_dataf)):
    for j in range(0, len(price_dataf)):
        aa = str(combine_dataf.Date.iloc[i])
        bb = ""
        for ii in range(0, len(aa)):
            if ii==3:
                bb = bb + aa[ii] + "-"
        
            elif ii==5:
                bb = bb + aa[ii] + "-"
            else:
                bb = bb + aa[ii]
        tweet_date = bb
        stock_date = str(price_dataf.Date.iloc[j])
#         print(str(tweet_date))
#         print(str(stock_date))
        if (str(stock_date) == str(tweet_date)): 
            combine_dataf.set_value(i, 'Prices', float(price_dataf.Close[j]))
            break
ccdata = combine_dataf
ccdata
```

## 5. fill the blank price data

```python
if(ccdata.Prices.iloc[0]==""):
    j=0
    while (ccdata.Prices.iloc[j]==""):
        j += 1
    ccdata.Prices.iloc[0] = ccdata.Prices.iloc[j]
for i in range(1,len(ccdata)):
    if(ccdata.Prices.iloc[i]==""):
        ccdata.Prices.iloc[i] = ccdata.Prices.iloc[i-1]
```

```python
ccdata["Comp"] = ''
ccdata["Negative"] = ''
ccdata["Neutral"] = ''
ccdata["Positive"] = ''
ccdata
```

## 6. do sentiment analysis

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import unicodedata
sentiment_i_a = SentimentIntensityAnalyzer()
for indexx, row in ccdata.T.iteritems():
    try:
        sentence_i = unicodedata.normalize('NFKD', ccdata.loc[indexx, 'Tweet'])
        sentence_sentiment = sentiment_i_a.polarity_scores(sentence_i)
        ccdata.set_value(indexx, 'Comp', sentence_sentiment['compound'])
        ccdata.set_value(indexx, 'Negative', sentence_sentiment['neg'])
        ccdata.set_value(indexx, 'Neutral', sentence_sentiment['neu'])
        ccdata.set_value(indexx, 'Positive', sentence_sentiment['pos'])
    except TypeError:
        print (stocks_dataf.loc[indexx, 'Tweet'])
        print (indexx)
ccdata
```

### plot the pie chart to show sentiment result, just for fun.

```python
posi=0
nega=0
for i in range (0,len(ccdata)):
    get_val=ccdata.Negative[i]
    if(float(get_val)<(0.077)):######## use neg as threshold
        nega=nega+1
    if(float(get_val>=(0.077))):
        posi=posi+1
posper=(posi/(len(ccdata)))*100
negper=(nega/(len(ccdata)))*100
print("% of positive tweets= ",posper)
print("% of negative tweets= ",negper)
arr=np.asarray([posper,negper], dtype=int)
mlpt.pie(arr,labels=['positive','negative'])
mlpt.plot()
```
